### Fonction

```python
def get_url_recettes (nom):
    """
    Prend un nom de recette en chaine de caractère en entrée et renvoit une liste des recettes sur Marmiton
    """
    requete_propre = nom.replace(" ", "-")
    url = f"https://www.marmiton.org/recettes/recherche.aspx?aqt={requete_propre}"
    
    page_marmiton = requests.get(url)
    html_marmiton = page_marmiton.text
    
    soup = BeautifulSoup(html_marmiton, "html.parser")
    
    liste_liens = soup.find_all("a", class_="MRTN__sc-1gofnyi-2 gACiYG")
    liste_url = []
    for lien in liste_liens:
        liste_url.append(lien.get("href"))
        
    index = 0
    for lien in liste_url:
        if lien[0] == "/":
            liste_url[index] = "https://www.marmiton.org"+lien
        index += 1
        
    index = 0
    url_reel = []
    for lien in liste_url:
        if lien.endswith("aspx") :
            url_reel.append(lien)
        index += 1
    return(url_reel)
```

```python
get_url_recettes("fondant au chocolat")
```

### Fonction

```python
def get_ingredient_url(url):
    """ 
    Prend un url en entrée et renvoit la liste des ingrédients de la recette
    """
    page_marmiton = requests.get(url)
    html_marmiton = page_marmiton.text
    
    soup = BeautifulSoup(html_marmiton, "html.parser")
    
    liste_ingrédients = []
    ingrédients = soup.find_all(class_="SHRD__sc-10plygc-0 kWuxfa")
    for nom in ingrédients:
        liste_ingrédients.append(nom.text)
    
    liste_proportion = []
    proportions = soup.find_all(class_="SHRD__sc-10plygc-0 epviYI")
    for prop in proportions:
        liste_proportion.append(prop.text.replace("\xa0", " "))
    
    index = 0
    liste = []
    for nom in liste_ingrédients:
        liste.append(liste_proportion[index]+" "+nom)
        index += 1
    return(liste)
    
```

```python
get_ingredient_url("https://www.marmiton.org/recettes/recette_ramequins-fondants-au-chocolat_15816.aspx")
```
